combinamos los dos df

```python
#correjimos los nombres de las columnas
penetracion_hogares_provincia.columns = penetracion_hogares_provincia.columns.str.strip()
penetracion_poblacion_provincia.columns = penetracion_poblacion_provincia.columns.str.strip()

df_combinados_penetracion = pd.merge(penetracion_hogares_provincia, penetracion_poblacion_provincia, on=['Año', 'Trimestre', 'Provincia'])
df_combinados_penetracion.info()
df_combinados_penetracion.sample(4)
```

Revisamos el nuevo df combinado

```python
df_combinados_penetracion.sample(5)
```

```python
#revicion de df 
acceso_por_trecnologia_provincia.info()
acceso_por_trecnologia_provincia.sample(10)
```

```python
# Eliminar asterisco (*) y espacios extra
acceso_por_trecnologia_provincia['Trimestre'] = acceso_por_trecnologia_provincia['Trimestre'].str.replace('*', '', regex=False)

# Convertir la columna a entero
acceso_por_trecnologia_provincia['Trimestre'] = acceso_por_trecnologia_provincia['Trimestre'].astype(int)

df_combinados_penetracion['Provincia'] = df_combinados_penetracion['Provincia'].str.strip()
acceso_por_trecnologia_provincia['Provincia'] = acceso_por_trecnologia_provincia['Provincia'].str.strip()
acceso_por_trecnologia_provincia['Trimestre'] = acceso_por_trecnologia_provincia['Trimestre'].astype(int)
df_combinados_penetracion['Trimestre'] = df_combinados_penetracion['Trimestre'].astype(int)

# Si no es string, conviértelo a string
if acceso_por_trecnologia_provincia['Año'].dtype != 'object':
    acceso_por_trecnologia_provincia['Año'] = acceso_por_trecnologia_provincia['Año'].astype(str)

# Extrae solo los caracteres numéricos
acceso_por_trecnologia_provincia['Año'] = acceso_por_trecnologia_provincia['Año'].str.extract('(\d+)', expand=False)

# Convierte la columna a entero
acceso_por_trecnologia_provincia['Año'] = acceso_por_trecnologia_provincia['Año'].astype(int)

df_combinados= pd.merge(df_combinados_penetracion,acceso_por_trecnologia_provincia, on=['Año', 'Trimestre', 'Provincia'], how= 'outer',indicator=True)
df_combinados.info()
df_combinados.head()
```

```python
# Convertir los nombres de las columnas a minúsculas
df_combinados.columns = df_combinados.columns.str.lower()

# Verificar los nuevos nombres de las columnas
print(df_combinados.columns)
```

limpieza a las columnas clave para garantizar que los valores coincidan correctamente

```python
print("Valores únicos en 'provincia' de df_combinados:")
print(df_combinados['provincia'].unique())
print("\nValores únicos en 'provincia' de dial_baf_provincia:")
print(dial_baf_provincia['Provincia'].unique())

print("\nValores únicos en 'trimestre' de df_combinados:")
print(df_combinados['trimestre'].unique())
print("\nValores únicos en 'trimestre' de dial_baf_provincia:")
print(dial_baf_provincia['Trimestre'].unique())

print("\nValores únicos en 'año' de df_combinados:")
print(df_combinados['año'].unique())
print("\nValores únicos en 'año' de dial_baf_provincia:")
print(dial_baf_provincia['Año'].unique())
```

```python
# Normalizar nombres de columnas en ambos DataFrames
for df in [df_combinados, dial_baf_provincia]:
    df.columns = df.columns.str.strip().str.lower()

# Verificar si 'provincia' ahora está disponible
print(df_combinados.columns)
print(dial_baf_provincia.columns)

dial_baf_provincia.columns = dial_baf_provincia.columns.str.strip().str.lower()

# Convertir la columna 'trimestre' a tipo entero en ambos DataFrames
dial_baf_provincia['trimestre'] = pd.to_numeric(dial_baf_provincia['trimestre'], errors='coerce')
df_combinados['trimestre'] = pd.to_numeric(df_combinados['trimestre'], errors='coerce')

df_combinados2 = pd.merge(dial_baf_provincia,df_combinados,on=['año', 'trimestre', 'provincia'], how='outer')
df_combinados2.sample(6)
```
